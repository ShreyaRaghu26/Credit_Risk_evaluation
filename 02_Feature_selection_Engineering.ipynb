```python
#comeback to this 
def reduce_memory_usage(df):
    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = df[col].astype('float32')
    for col in df.select_dtypes(include=['int64']).columns:
        df[col] = df[col].astype('int32')
    return df

def aggregate_numerical_feature(df, feature, ref_date, customer_col='customer_ID', date_col='date'):
    windows = [3, 6, 9, 12]
    agg_list = []

    for window in windows:
        mask = (df[date_col] <= ref_date) & (df[date_col] > (ref_date - pd.DateOffset(months=window)))
        temp = df.loc[mask, [customer_col, feature]]

        grouped = temp.groupby(customer_col)[feature].agg([
            (f'{feature}Ave{window}', 'mean'),
            (f'{feature}Min{window}', 'min'),
            (f'{feature}Max{window}', 'max'),
            (f'{feature}Sum{window}', 'sum')
        ]).reset_index()

        grouped = reduce_memory_usage(grouped)
        agg_list.append(grouped)

    # Safe merging using loop (avoids deep recursion and memory spikes)
    df_agg = agg_list[0]
    for next_df in agg_list[1:]:
        df_agg = pd.merge(df_agg, next_df, on=customer_col, how='outer')

    # Latest value as of ref_date (April 2018)
    latest = df[df[date_col] == ref_date][[customer_col, feature]].rename(
        columns={feature: f'{feature}Apr{ref_date.year}'}
    )
    df_agg = pd.merge(df_agg, latest, on=customer_col, how='left')

    # Value from previous April
    prev_year_date = ref_date - pd.DateOffset(years=1)
    prev_year = df[df[date_col] == prev_year_date][[customer_col, feature]].rename(
        columns={feature: f'{feature}Apr{prev_year_date.year}'}
    )
    df_agg = pd.merge(df_agg, prev_year, on=customer_col, how='left')

    return reduce_memory_usage(df_agg)
```

```python
def aggregate_one_hot_feature(df, feature, ref_date, customer_col='customer_ID', date_col='date'):
  
    # Define time windows in months
    windows = [6, 12]
    agg_list = []

    for window in windows:
        # Filter data within the time window
        mask = (df[date_col] <= ref_date) & (df[date_col] > (ref_date - pd.DateOffset(months=window)))

        # Only keep the necessary columns to reduce memory
        temp = df.loc[mask, [customer_col, feature]]

        # Group by customer and perform aggregations
        grouped = temp.groupby(customer_col)[feature].agg([
            (f'{feature}Response_Rate{window}', 'mean'),  # % of times it is 1 (response rate)
            (f'{feature}Ever_Response{window}', 'max')    # Was it ever 1?
        ]).reset_index()

        agg_list.append(grouped)

    # Merge all aggregations on customer_id
    if agg_list:
        df_agg = reduce(lambda left, right: pd.merge(left, right, on=customer_col, how='outer'), agg_list)
    else:
        df_agg = pd.DataFrame(columns=[customer_col])

    # Get the value for the feature as of ref_date
    latest = df[df[date_col] == ref_date][[customer_col, feature]].rename(
        columns={feature: f'{feature}Apr{ref_date.year}'}
    )

    # Merge latest value into the final aggregated dataframe
    df_final = pd.merge(df_agg, latest, on=customer_col, how='left')

    return df_final
```

```python
one_hot_encoded_cols = [col for col in train_date_filtered.columns if any(col.startswith(f"{prefix}_") for prefix in cat_feature)]
```

```python
# Helper function to reduce memory usage
def reduce_memory_usage(df):
    for col in df.select_dtypes(include=['float64']).columns:
        df[col] = df[col].astype('float32')
    for col in df.select_dtypes(include=['int64']).columns:
        df[col] = df[col].astype('int32')
    for col in df.select_dtypes(include=['object']).columns:
        if df[col].nunique() < df.shape[0] // 2:
            df[col] = df[col].astype('category')
    return df

# Aggregation for categorical (one-hot encoded) features
categorical_agg_list = []

for feature in one_hot_encoded_cols:
    print(f'Aggregating one-hot encoded feature: {feature}')
    agg_df = aggregate_one_hot_feature(train_date_filtered, feature, april_2018,
                                       customer_col='customer_ID', date_col='date')
    agg_df = reduce_memory_usage(agg_df)
    categorical_agg_list.append(agg_df)

# Safe merging using a loop instead of reduce
if categorical_agg_list:
    df_categorical_features = categorical_agg_list[0]
    for next_df in categorical_agg_list[1:]:
        df_categorical_features = pd.merge(df_categorical_features, next_df, on='customer_ID', how='outer')
else:
    df_categorical_features = pd.DataFrame()
```

```python
# Initialize output list
numerical_agg_list = []

# Loop through all numerical columns
for feature in numerical_cols:
    print(f'Aggregating numerical feature: {feature}')
    agg_df = aggregate_numerical_feature(train_date_filtered, feature, april_2018,
                                         customer_col='customer_ID', date_col='date')
    numerical_agg_list.append(agg_df)

# Merge all aggregated numerical features
if numerical_agg_list:
    df_numerical_features = numerical_agg_list[0]
    for next_df in numerical_agg_list[1:]:
        df_numerical_features = pd.merge(df_numerical_features, next_df, on='customer_ID', how='outer')
else:
    df_numerical_features =pd.DataFrame()
```

```python
# Ensure both feature sets exist
if 'df_numerical_features' not in locals():
    df_numerical_features = pd.DataFrame()

if 'df_categorical_features' not in locals():
    df_categorical_features = pd.DataFrame()

# Merge logic
if not df_numerical_features.empty and not df_categorical_features.empty:
    print("Merging numerical and categorical features...")
    df_final_features = pd.merge(
        df_numerical_features,
        df_categorical_features,
        on='customer_ID',
        how='outer'
    )
elif not df_numerical_features.empty:
    print("Using only numerical features.")
    df_final_features = df_numerical_features.copy()
elif not df_categorical_features.empty:
    print("Using only categorical features.")
    df_final_features = df_categorical_features.copy()
else:
    print("No features available. Creating empty DataFrame.")
    df_final_features = pd.DataFrame()

# OPTIONAL: Merge the target column
if 'train_date_filtered' in locals() and 'target' and 'date' in train_date_filtered.columns:
    print("Merging target column...")
    target_df = train_date_filtered[['customer_ID', 'target','date']].drop_duplicates(subset='customer_ID')
    df_final_features = pd.merge(df_final_features, target_df, on='customer_ID', how='left')
else:
    print("Warning: Could not find 'target' column in train_date_filtered.")

# Handle missing values safely, even for categorical columns
for col in df_final_features.columns:
    col_data = df_final_features[col]
    
    if pd.api.types.is_categorical_dtype(col_data):
        # Get the category type to match the fill value
        if len(col_data.cat.categories) > 0:
            cat_type = type(col_data.cat.categories[0])
            fill_value = cat_type(0)  # Convert 0 to the same type
        else:
            fill_value = 0  # fallback

        # Add category if it's not there
        if fill_value not in col_data.cat.categories:
            try:
                df_final_features[col] = col_data.cat.add_categories([fill_value])
            except ValueError:
                pass  # Already present or incompatible type

        # Now fill safely
        df_final_features[col] = df_final_features[col].fillna(fill_value)
    else:
        df_final_features[col] = col_data.fillna(0)

# Final preview
print("Final features shape:", df_final_features.shape)
df_final_features.head()
```

```python
import pandas as pd
import re

# Use your actual DataFrame name here — should be the one after one-hot encoding
# Example: df_encoded = pd.get_dummies(df, columns=categorical_columns)
# Make sure df_encoded is already defined in your code

# Define feature category patterns based on naming conventions
category_patterns = {
    'Delinquency': r'^D_',
    'Balance': r'^B_',
    'Risk': r'^R_',
    'Spend': r'^S_',
    'Payment': r'^P_',
    'Date': r'^S_2$',  # Specific for timestamp column
    'Customer Info': r'^customer_ID$',
    'Unknown/Other': r'^.*'  # Catch-all fallback
}

# Initialize dictionary to hold counts
category_counts = {cat: 0 for cat in category_patterns}

# Categorize and count columns
for col in df_final_features.columns:
    matched = False
    for category, pattern in category_patterns.items():
        if re.match(pattern, col):
            category_counts[category] += 1
            matched = True
            break
    if not matched:
        category_counts['Unknown/Other'] += 1

# Convert result to DataFrame for display or export
category_summary = pd.DataFrame.from_dict(category_counts, orient='index', columns=['# Columns'])
category_summary.index.name = 'Feature Category'

# Display the result
print(category_summary)

# Optional: Save to CSV
# category_summary.to_csv('feature_category_counts.csv')
```
